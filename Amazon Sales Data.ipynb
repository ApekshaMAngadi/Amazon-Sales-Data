```python
import matplotlib.pyplot as plt
import seaborn as sns

# Month-wise Sales Trend
month_sales = diwali_sales_data.groupby('Registration_Month')['Amount'].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x=month_sales.index, y=month_sales.values)
plt.title('Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()

# Year-wise Sales Trend
year_sales = diwali_sales_data.groupby('Registration_Year')['Amount'].sum()
plt.figure(figsize=(12, 6))
sns.barplot(x=year_sales.index, y=year_sales.values)
plt.title('Year-wise Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()

# Yearly-Month-wise Sales Trend
year_month_sales = diwali_sales_data.groupby(['Registration_Year', 'Registration_Month'])['Amount'].sum().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(year_month_sales, cmap='viridis')
plt.title('Yearly-Month-wise Sales Trend')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```
